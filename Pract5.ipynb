```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df_2020 = pd.read_csv('IT Salary Survey EU  2020.csv')
df_2020.head(2)
```

```python
df_2018 = pd.read_csv('IT Salary Survey EU 2018.csv')
df_2018.head(2)
```

```python
df_2019=pd.read_csv('T Salary Survey EU 2019.csv')
df_2019.head(10)
```

```python
df = pd.merge(df_2018,df_2019,on='Age')
df = pd.merge(df,df_2020,on='Age')
```

```python
table = pd.pivot_table(data=df,index='Gender',values='Age',aggfunc=np.sum)
table
```

```python
#plotting a bar graph.
#bar graph
plt.bar(table.index,table['Age'])

#xticks 
plt.xticks(rotation=70) 

#x-axis labels 
plt.xlabel('Gender') 

#y-axis labels 
plt.ylabel('Age') 

#plot title 
plt.title('Age Ratio') 

#save plot 
plt.savefig('bargraph.png' ,dpi=300, bbox_inches='tight') 

#displayplot
plt.show();
```
