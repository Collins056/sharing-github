```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('IT Salary Survey EU  2020.csv')
df.head()
```

```python
 df.shape
```

```python
#this dataset comprises of 1253 observations and 23 characteristics
```

```python
sns.heatmap(df.isnull(),cbar=False,yticklabels=False,cmap = 'viridis')
```

```python
#this dataset has missing values.
#this is shown by the different colour shades in the purple background.
```

```python
#To check correlation
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(),cmap='Blues',annot=False)
```

```python
#dark shades represents positive correlation while lighter shades represents negative correlation.
```

```python
#Age correlation matrix
k = 12 #number of variables for heatmap
cols = df.corr().nlargest(k,'Age')['Age'].index
cm = df[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```
